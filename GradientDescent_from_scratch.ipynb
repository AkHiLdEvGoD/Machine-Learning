```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_regression(n_samples=100,n_features=1,n_targets=1,n_informative=1,noise=20,random_state=42)
plt.scatter(X,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
print('intercept :',lr.intercept_)
print('slope :',lr.coef_)
```

```python
class gradient_descent:
    
    def __init__(self,learning_rate,epochs):
        self.lr=learning_rate
        self.epochs=epochs
        self.m=-430
        self.b=754
        self.b_count = []
        self.m_count = []

    def fit(self,X,y):
        for i in range(self.epochs):
            slope_b = -2 * np.sum(y - self.m*X.ravel() - self.b)
            slope_m = -2 * np.sum((y-self.m*X.ravel() - self.b)*X.ravel())

            self.b = self.b - self.lr*slope_b
            self.b_count.append(self.b) 
            self.m = self.m - self.lr*slope_m                        
            self.m_count.append(self.m) 


        print('intercept :',self.b)
        print('slope :',self.m)

    def plot(self):
        fig,[ax1,ax2] = plt.subplots(1,2,figsize=(12,5))
        
        ax1.plot(list(range(self.epochs)),self.b_count)
        ax1.set_title('Convergence of intercept b w.r.t no_of_epochs')
        ax1.set_ylabel('Intercept(b)')
        ax1.set_xlabel('No. of epochs')

        ax2.plot(list(range(self.epochs)),self.m_count)
        ax2.set_title('Convergence of weight m w.r.t no_of_epochs')
        ax2.set_ylabel('Weight(w)')
        ax2.set_xlabel('No. of epochs')

    def plot_best_fit_line(self,X,y):
        plt.scatter(X,y)
        plt.plot(X,self.m*X+self.b,'r-')
        plt.title('Best Fit line')

    # def predict(self,X):
    #     return self.m*X + self.b
```

```python
gd = gradient_descent(0.001,50)
gd.fit(X,y)
```

```python
gd.plot()
```

```python
gd.plot_best_fit_line(X,y)
```
